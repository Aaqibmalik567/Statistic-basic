

**Q1. What is statistics, and why is it important?**
A: Statistics is the science of collecting, organizing, analyzing, and interpreting data. It helps make decisions and predictions based on data.

---

**Q2. What are the two main types of statistics?**
A:

1. Descriptive Statistics
2. Inferential Statistics

---

**Q3. What are descriptive statistics?**
A: These summarize and describe the features of a dataset using tools like mean, median, mode, range, and standard deviation.

---

**Q4. What is inferential statistics?**
A: It uses a sample of data to make predictions or inferences about a population.

---

**Q5. What is sampling in statistics?**
A: Sampling is the process of selecting a subset of individuals from a population to estimate the characteristics of the whole group.

---

**Q6. What are the different types of sampling methods?**
A:

* Random Sampling
* Stratified Sampling
* Systematic Sampling
* Cluster Sampling
* Convenience Sampling

---

**Q7. What is the difference between random and non-random sampling?**
A: Random sampling gives every item an equal chance of selection, while non-random sampling involves personal judgment or convenience.

---

**Q8. Define and give examples of qualitative and quantitative data.**
A:

* Qualitative (Categorical): Gender, color, city
* Quantitative (Numerical): Height, age, income

---

**Q9. What are the different types of data in statistics?**
A:

* Nominal
* Ordinal
* Interval
* Ratio

---

**Q10. Explain nominal, ordinal, interval, and ratio levels of measurement.**
A:

* Nominal: Categories without order (e.g., gender)
* Ordinal: Ordered categories (e.g., rankings)
* Interval: Ordered, with equal spacing, no true zero (e.g., temperature)
* Ratio: Like interval, but has a true zero (e.g., weight, income)

---

**Q11. What is the measure of central tendency?**
A: It is a single value that represents the center or average of a dataset. (Mean, Median, Mode)

---

**Q12. Define mean, median, and mode.**
A:

* Mean: Arithmetic average
* Median: Middle value when sorted
* Mode: Most frequently occurring value

---

**Q13. What is the significance of the measure of central tendency?**
A: It helps summarize a large dataset with a single representative value.

---

**Q14. What is variance, and how is it calculated?**
A: Variance is the average of the squared differences from the mean. It shows how data is spread out.

---

**Q15. What is standard deviation, and why is it important?**
A: Standard deviation is the square root of variance. It is widely used to measure data spread in the same unit as the data.

---

**Q16. Define and explain the term range in statistics.**
A: Range is the difference between the highest and lowest values in a dataset.

---

**Q17. What is the difference between variance and standard deviation?**
A: Variance is in squared units; standard deviation is its square root, making it easier to interpret.

---

**Q18. What is skewness in a dataset?**
A: Skewness describes the asymmetry of a data distribution.

---

**Q19. What does it mean if a dataset is positively or negatively skewed?**
A:

* Positive skew: Right tail is longer
* Negative skew: Left tail is longer

---

**Q20. Define and explain kurtosis.**
A: Kurtosis measures the ‚Äútailedness‚Äù or peakedness of a distribution compared to a normal distribution.

---

**Q21. What is the purpose of covariance?**
A: Covariance measures how two variables change together (positive or negative relationship).

---

**Q22. What does correlation measure in statistics?**
A: Correlation measures the strength and direction of the linear relationship between two variables.

---

**Q23. What is the difference between covariance and correlation?**
A: Correlation is standardized (‚àí1 to +1), while covariance is not and depends on the units of measurement.

---

**Q24. What are some real-world applications of statistics?**
A:

* Healthcare (disease prediction)
* Business (market analysis)
* Sports (performance analysis)
* Government (population studies)
* Finance (stock market forecasting)


```python
# üìå Q1: How do you calculate the mean, median, and mode of a dataset?

import numpy as np
import pandas as pd

# Sample dataset
data = np.array([12, 15, 14, 10, 8, 12, 14, 16, 18, 20])

# Calculating Mean
mean = np.mean(data)

# Calculating Median
median = np.median(data)

# Calculating Mode using pandas
mode = pd.Series(data).mode().iloc[0]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# üìå Q2: Write a Python program to compute the variance and standard deviation of a dataset

import numpy as np

data = np.array([12, 15, 14, 10, 8, 12, 14, 16, 18, 20])

# Variance
variance = np.var(data)

# Standard Deviation
std_dev = np.std(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# üìå Q3: Write a Python function to calculate the range of a dataset

def calc_range(dataset):
    return np.max(dataset) - np.min(dataset)

print("Range:", calc_range(data))
```

```python
# üìå Q5: Create a dataset and plot its histogram to visualize skewness

import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Sample dataset (same as before)
sns.histplot(data, kde=True)
plt.title("Histogram with KDE - Skewness")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))
```

```python
# üìå Q6: Generate a dataset and demonstrate positive and negative skewness

positive_skew = np.random.exponential(size=1000)
negative_skew = -np.random.exponential(size=1000)

plt.figure(figsize=(12,5))

# Positive Skew
plt.subplot(1, 2, 1)
sns.histplot(positive_skew, kde=True, color='blue')
plt.title("Positive Skew")

# Negative Skew
plt.subplot(1, 2, 2)
sns.histplot(negative_skew, kde=True, color='red')
plt.title("Negative Skew")

plt.tight_layout()
plt.show()
```

```python
# üìå Q7: Write a Python script to calculate covariance between two datasets

x = np.random.randint(10, 100, 50)
y = x + np.random.randint(-10, 10, 50)

cov_matrix = np.cov(x, y)
print("Covariance between x and y:", cov_matrix[0, 1])
```

```python
# üìå Q8: Write a Python script to calculate the correlation coefficient between two datasets

corr_matrix = np.corrcoef(x, y)
print("Correlation Coefficient between x and y:", corr_matrix[0, 1])
```

```python
# üìå Q10: Implement and compare simple random sampling and stratified sampling

import pandas as pd

df = pd.DataFrame({
    'gender': ['M', 'F', 'M', 'F', 'F', 'M', 'M', 'F', 'M', 'F'],
    'score': [88, 92, 75, 65, 78, 90, 85, 70, 95, 60]
})

# Simple Random Sampling
random_sample = df.sample(n=4)

# Stratified Sampling (2 samples per gender)
stratified_sample = df.groupby('gender', group_keys=False).apply(lambda x: x.sample(2))

print("Random Sample:\n", random_sample)
print("\nStratified Sample:\n", stratified_sample)
```

```python
# üìå Q11: Calculate the mean, median, and mode of grouped data

grouped_data = pd.DataFrame({
    'Class Interval': ['0-10', '10-20', '20-30'],
    'Frequency': [5, 8, 7]
})

grouped_data['Midpoint'] = [5, 15, 25]

# Mean for grouped data
mean_grouped = np.average(grouped_data['Midpoint'], weights=grouped_data['Frequency'])

print("Grouped Data Mean:", mean_grouped)
```
